```python

# Generate fake temperature data (30 days)
days = np.arange(1, 31)
temps = 25 + 5 * np.sin(days/3) + np.random.randn(30)  # pattern + noise

# Plot
plt.plot(days, temps, marker='o', linestyle='--', color='b')
plt.title("Daily Temperatures (Simulated)")
plt.xlabel("Day")
plt.ylabel("Temperature (°C)")
plt.show()

# Quick stats with NumPy
print("Average Temp:", np.mean(temps))
print("Max Temp:", np.max(temps))
print("Min Temp:", np.min(temps))
```
